## Key problems solved by LaMa


Most of the machine learning project about images is based on convolution. Since the convolution give the AI ability to get the relationship between a pixel and other pixels. However, the convolution is
- very performance consuming
- based on pixels, and scaling will break the realtion ship a pixel with other pixels. Hence, the result will be wrong after convolution and scaling to high resilution image.

LaMa use Fourier Convolutions to make it be possible that the scaling will keep the result is correct.

## Fast Fourier Convolution (FFC)

### Fast Fourier Transform (FFT)
$$
A(x)=\sum_{i=0}^{n-1}a_ix^i
$$
$$
y_i=A(\omega_n^i)=\sum_{i=0}^{n-1}\omega_n^{ij}x^i
$$
$$
O(n^2)
$$
$$
A(x) = [a_0, a_1, a_2, \dots, a_{n-1}]'
$$
$$
A^{[0]}(x) = [a_0, a_2, a_4, \dots, a_{n-2}]'
$$
$$
A^{[1]}(x) = [a_1, a_3, a_5, \dots, a_{n-1}]'
$$
$$
A(x) = A^{[0]}(x^2) + xA^{[1]}(x^2)
$$
$$
\left\{\begin{aligned}
A\left(\omega_{n}^{k}\right) &=A^{[0]}\left(\omega_{\frac{n}{2}}^{k}\right)+\omega_{n}^{k} A^{[1]}\left(\omega_{\frac{n}{2}}^{k}\right) \\
A\left(\omega_{n}^{k+\frac{n}{2}}\right) &=A^{[0]}\left(\omega_{\frac{n}{2}}^{k}\right)-\omega_{n}^{k} A^{[1]}\left(\omega_{\frac{n}{2}}^{k}\right)
\end{aligned}\right.
$$
$$
O(n\log n)
$$
